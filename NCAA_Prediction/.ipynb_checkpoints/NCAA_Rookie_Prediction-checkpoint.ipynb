```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
NBA_stats_renamed_df = pd.read_csv("FINAL_NBA_Stats.csv")
NCAA_stats_renamed_df = pd.read_csv("FINAL_NCAA_Stats.csv")
```

```python
from sklearn.model_selection import train_test_split
variables = NCAA_stats_renamed_df.columns
for line in variables:
    print(line)
    X = NCAA_stats_renamed_df[line].values.reshape(-1, 1)
    y = NBA_stats_renamed_df[line].values.reshape(-1, 1)

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
    
    # Linear Regression
    reg = LinearRegression().fit(X_train, y_train)
    score = reg.score(X_train, y_train)
    print(f"Linear Regression Score: {score}")
    #prediction = reg.predict(np.array([[3, 5]]))
    
    # Random Forest
    regr = RandomForestRegressor(max_depth=2, random_state=0)
    regr.fit(X_train, y_train)

    training_score = regr.score(X_train, y_train)
    testing_score = regr.score(X_test, y_test)
    print(f"Random Forest Score: {training_score}")
    #print(f"Testing Score: {testing_score}")

#     plt.scatter(X_train, y_train,  color='black')
#     plt.title("Training Data")
#     plt.xlabel(f"NCAA {line}")
#     plt.ylabel(f"NBA {line}")
#     plt.show()

    print("--------------------")
```
