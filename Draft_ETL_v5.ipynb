```python
# Dependencies
import pandas as pd
```

## Extract CSV and XLS into DataFrames

```python
# NBA stats dataframe from .csv
path1 = "data/NBA 2020-2021 v3.csv"
nba_20_21_df = pd.read_csv(path1, encoding = 'utf8')  

nba_20_21_df.head()
```

```python
# NBA salaries from .csv
path2 = "data/NBA_Salaries.csv"
nba_salaries_df = pd.read_csv(path2, encoding = 'utf8')  

nba_salaries_df.head()
```

## TRANSFORM DataFrames

### TRANSFORM DataFrame


```python
# Column rename for salaries
nba_salaries_df = nba_salaries_df.rename(columns = {'2021/22': 'Salary'})
nba_salaries_df.head()
```

```python
# Get last team for any player even in repeated records (There are season transfers)
teams_20_21_df = nba_20_21_df.drop_duplicates("Player", keep='last')
teams_20_21_df
```

```python
# keep specific columns needed in the future merge for initial fields
teams_20_21_df = teams_20_21_df[['Player', 'Pos', 'Age', 'Tm']]
teams_20_21_df.head()
```

```python
# review previous upload from .csv
nba_20_21_df.head()
```

```python
# Replace fields with NaN to zero. 
# It was produced by obtaining an infinite value in calculated fields
nba_20_21_df = nba_20_21_df.fillna(0)
nba_20_21_df.head()
```

```python
# Check columns
for col in nba_20_21_df.columns:
    print(col)
```

```python
# Group by and sum numeric fields in repeated players
# transferred during season
sum_20_21_df = nba_20_21_df.groupby(['Player']).sum()
sum_20_21_df.head()
```

```python
# Option to print full rows and columns
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
# calculate fields with added values with respective formula
sum_20_21_df['FG%'] = sum_20_21_df['FG'] / sum_20_21_df['FGA']
sum_20_21_df['3P%'] = sum_20_21_df['3P'] / sum_20_21_df['3PA']

# eFG% - Effective Field Goal Percentage; the formula is (FG + 0.5 * 3P) / FGA. 
# This statistic adjusts for the fact that a 3-point field goal is worth one more point than a 2-point field goal. 
sum_20_21_df['eFG%'] = (sum_20_21_df['FG'] + 0.5 * sum_20_21_df['3P']) / sum_20_21_df['FGA']
sum_20_21_df['FT%'] = sum_20_21_df['FT'] / sum_20_21_df['FTA']

sum_20_21_df
```

```python
sum_20_21_df = sum_20_21_df.drop(columns=['Age', 'Rk'])
sum_20_21_df
```

```python
# Replace NaN with '0'.
sum_20_21_df = sum_20_21_df.fillna(0)
sum_20_21_df.head()
```

```python
# merge player/tems with stats datasets
merge_1_df = teams_20_21_df.merge(sum_20_21_df, how ='left', on="Player")
merge_1_df.head()
```

```python
# Merge salaries
merge_2_df = merge_1_df.merge(nba_salaries_df, how ='left', on="Player")
merge_2_df.head()
```

```python
# Drop unnecessary column 'ID'
merge_2_df = merge_2_df.drop(columns=['ID'])
merge_2_df.head()
```

```python
# Check salaries are updated
merge_2_df.loc[merge_2_df["Salary"].isnull()]
```

## LOADING Phase

```python
merge_2_df.to_csv('data/FINAL.csv', index=False)
```
